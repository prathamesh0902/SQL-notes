### Staging Layer

```sql
CREATE TABLE dwhouse.stg_sales
AS
SELECT * FROM dwhouse.orders
```

### Transformation

```sql
CREATE VIEW dwhouse.trans_sales
AS
select * from dwhouse.stg_sales where Country = 'USA'
```

```sql
select * from dwhouse.trans_sales
```

### Core Layer

```sql
CREATE OR REPLACE TABLE dwhouse.core_sales (
  orderID INT, orderDate DATE, customerID INT, customerName STRING, customerEmail STRING, ProductID INT, ProductName STRING, Quantity INT, UnitPrice DECIMAL(10,2), RegionID INT, RegionName STRING, Country STRING, totalAmount DECIMAL(10,2)
);
```

```sql
INSERT INTO dwhouse.core_sales
SELECT * FROM dwhouse.trans_sales
```
