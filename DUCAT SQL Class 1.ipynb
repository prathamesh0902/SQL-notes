
## Schema
- A **schema** is the blueprint or design of a database.
- It defines the **connection of tables** and provides an **overview** of database structure.

---

## Components of a Database (4 Key Elements)

1. **Tables**
   - Organized in **rows (records)** and **columns (fields/attributes)**.
   - Represents the fundamental storage of data.
   - Relations exist between tables to maintain data integrity.

2. **Views**
   - A **mirror image of data** derived from one or more tables.
   - Provides a way to look at specific data without duplicating storage.
   - Useful for security, abstraction, and simplified querying.

3. **Stored Procedures**
   - Predefined and reusable SQL code blocks.
   - Used to encapsulate logic like **insert, update, delete, or complex queries**.
   - Improves performance and reduces redundancy.

4. **Functions**
   - Similar to stored procedures but **return a single value or table**.
   - Can be used directly in SQL queries.
   - Ideal for encapsulating reusable logic or calculations.
