# DQL (Data Query Language)

## Inserting Data

Importing data from `create-databases.sql` file

```sql
DROP DATABASE IF EXISTS sql_invoicing CASCADE;
CREATE DATABASE sql_invoicing;
USE sql_invoicing;

CREATE TABLE payment_methods (
  payment_method_id TINYINT NOT NULL,
  name STRING NOT NULL,
  PRIMARY KEY (payment_method_id)
);

CREATE TABLE clients (
  client_id INT NOT NULL,
  name STRING NOT NULL,
  address STRING NOT NULL,
  city STRING NOT NULL,
  state STRING NOT NULL,
  phone STRING,
  PRIMARY KEY (client_id)
);

CREATE TABLE invoices (
  invoice_id INT NOT NULL,
  number STRING,
  client_id INT,
  invoice_total DECIMAL(9,2),
  payment_total DECIMAL(9,2),
  invoice_date DATE,
  due_date DATE,
  payment_date DATE,
  PRIMARY KEY (invoice_id)
);

CREATE TABLE payments (
  payment_id INT NOT NULL,
  client_id INT NOT NULL,
  invoice_id INT NOT NULL,
  date DATE NOT NULL,
  amount DECIMAL(9,2) NOT NULL,
  payment_method TINYINT NOT NULL,
  PRIMARY KEY (payment_id)
);
```

```sql
DROP DATABASE IF EXISTS sql_store CASCADE;
CREATE DATABASE sql_store;
USE sql_store;

CREATE TABLE products (
  product_id INT NOT NULL,
  name STRING NOT NULL,
  quantity_in_stock INT NOT NULL,
  unit_price DECIMAL(4,2) NOT NULL,
  PRIMARY KEY (product_id)
);

CREATE TABLE shippers (
  shipper_id SMALLINT NOT NULL,
  name STRING NOT NULL,
  PRIMARY KEY (shipper_id)
);

CREATE TABLE customers (
  customer_id INT NOT NULL,
  first_name STRING NOT NULL,
  last_name STRING NOT NULL,
  birth_date DATE,
  phone STRING,
  address STRING NOT NULL,
  city STRING NOT NULL,
  state STRING NOT NULL,
  points INT NOT NULL,
  PRIMARY KEY (customer_id)
);

CREATE TABLE order_statuses (
  order_status_id TINYINT NOT NULL,
  name STRING NOT NULL,
  PRIMARY KEY (order_status_id)
);

CREATE TABLE orders (
  order_id INT NOT NULL,
  customer_id INT NOT NULL,
  order_date DATE NOT NULL,
  status TINYINT NOT NULL,
  comments STRING,
  shipped_date DATE,
  shipper_id SMALLINT,
  PRIMARY KEY (order_id)
);

CREATE TABLE order_items (
  order_id INT NOT NULL,
  product_id INT NOT NULL,
  quantity INT NOT NULL,
  unit_price DECIMAL(4,2) NOT NULL,
  PRIMARY KEY (order_id, product_id)
);

CREATE TABLE order_item_notes (
  note_id INT NOT NULL,
  order_id INT NOT NULL,
  product_id INT NOT NULL,
  note STRING NOT NULL,
  PRIMARY KEY (note_id)
);
```

```sql
DROP DATABASE IF EXISTS sql_hr CASCADE;
CREATE DATABASE sql_hr;
USE sql_hr;

CREATE TABLE offices (
  office_id INT NOT NULL,
  address STRING NOT NULL,
  city STRING NOT NULL,
  state STRING NOT NULL,
  PRIMARY KEY (office_id)
);

CREATE TABLE employees (
  employee_id INT NOT NULL,
  first_name STRING NOT NULL,
  last_name STRING NOT NULL,
  job_title STRING NOT NULL,
  salary INT NOT NULL,
  reports_to INT,
  office_id INT NOT NULL,
  PRIMARY KEY (employee_id)
  -- Foreign key constraints are not enforced in Databricks SQL, but keys are defined for reference
);
```

```sql
DROP DATABASE IF EXISTS sql_inventory CASCADE;
CREATE DATABASE sql_inventory;
USE sql_inventory;

CREATE TABLE products (
  product_id INT NOT NULL,
  name STRING NOT NULL,
  quantity_in_stock INT NOT NULL,
  unit_price DECIMAL(4,2) NOT NULL,
  PRIMARY KEY (product_id)
);
```

**Perform simple operations**


```sql
-- decrement 10 % pointunitprice * quantity = totalprice

select *, unit_price * quantity as Total_price
from order_items
```

```sql
-- concat first name- last name

select customer_id,
  concat(first_name, ' ', last_name) as Full_name,
  phone,
  address,
  points,
  points * 0.1 as increment_point
from customers;
```

### Time Intelligence

```sql
select birth_date,
  year(birth_date) as birth_year,
  monthname(birth_date) as birth_month,
  day(birth_date) as day_of_month,
  dayofweek(birth_date) as week_days,
  dayname(birth_date) as day_name,
  dayofyear(birth_date) as day_of_year,
  weekofyear(birth_date) as week_of_year
from
  customers;
```

```sql
select
    customer_id,
    concat(first_name, ' ', last_name) as Full_name,
    birth_date,
    now(),
    datediff(current_date, birth_date) as age_in_days,
    datediff(current_date, birth_date)/365 year_age_decimal,
    round(datediff(current_date, birth_date)/365, 1) as age_in_years,
    timestampdiff(year, birth_date, current_date) as Ages,
    timestampdiff(month, birth_date, current_date) as month_Ages
from
customers;
```

```sql
select
    customer_id,
    concat(first_name, ' ', last_name) as Full_name,
    birth_date,
    add_months(birth_date, 8) as New_date,
    year(current_date) - year(birth_date) + 1 as age_in_years,
    timestampdiff(year, birth_date, current_date) as Ages,
    timestampdiff(month, birth_date, current_date) as month_Ages,
    timestampdiff(day, birth_date)

from
customers
```

```sql
select
    customer_id,
    concat(first_name, ' ', last_name) as Full_name,
    birth_date,
    timestampdiff(year, birth_date, current_date) as Ages,
    timestampdiff(month, birth_date, current_date) % 12 as month_Ages,
    timestampdiff(day, birth_date, current_date) - (timestampdiff(month, birth_date, current_date)) as day_Ages

from
customers
```

```sql
-- Add days to a date
SELECT birth_date,
       date_add(birth_date, 365 * 2) AS plus_2_years_approx,
       birth_date + INTERVAL 2 YEAR AS plus_2_years_exact,
       date_add(birth_date, 10) AS plus_10_days,
       birth_date + INTERVAL 10 DAY AS plus_10_days_alt
FROM customers;
```
