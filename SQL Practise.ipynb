## Nested Queries

All employees whose salary is greater than the average salary of their own department.

```sql
-- Employees with salary > avg salary of dept

SELECT emp_id, emp_name, salary
FROM employee
WHERE salary > (
    SELECT AVG(salary)
    FROM employee e2
    WHERE e2.dept_id = employee.dept_id
)
ORDER BY salary DESC;
```
