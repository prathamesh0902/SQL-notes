```sql
-- stored procedure
-- parameters in, out, in and out

CREATE PROCEDURE 'info' ()
BEGIN
select c.customer_id,
	concat(first_name, " ", last_name) as Full_Name,
	o.order_id,
	order_date,
	shipped_date,
	unit_price * quantity as Total_Price
	address,
	phone
from orders o
join customers c
using(customer_id)
join order_items oi
using(order_id);
END


call info;
```
