```sql
-- left(col_name, no_of char)
-- mid(col_name, where, no_of char) 

    -- locate(" ",address) space_pos_no,
    -- locate(" ",address,locate(" ",address)+1) as 2nd_pos
    -- length(address) as total_no_char,
    -- locate(" ",address,locate(" ",address)+1) as 2nd_pos

use sql_store ;

select customer_id,
	concat(first_name," ",last_name) as Full_Name,
	address,
    left(address,locate(" ",address)-1) as Zipcode,
    mid(address, locate(" ",address)+1,
    locate(" ",address,locate(" ",address)+1) - locate(" ",address)) as State,
    right(address, length(address)  -  locate(" ",address,locate(" ",address)+1)) as State
from customers

-- where 
-- Relational Operator (<,>,<=,>=,==,!=) 
-- Logical Operator(and , or , not) 



select customer_id,
	concat(first_name," ",last_name) as Full_Name,
    address,
    birth_date,
    points
from customers
where not(year(birth_date) < 1986  or  points > 2000 and customer_id > 5)



select * 
from customers
where year(birth_date) >=1980 and year(birth_date) < 1990


select * , "Birthday mr's/miss's" as Status
from customers
where 
month(birth_date) = month(current_date) 
-- day(birth_date) = day(current_date) 


select *
from customers 
where customer_id in (
					select customer_id
					from orders 
					where year(order_date) = 2018)





-- between
-- membership operator (in , not in)
-- identity Operator(is , is not) 


select *
from customers
where state not in ('MA','VA','GA','CA')

where year(birth_date) between 1970 and 1991



select *
from customers
where (points between 1000 and 2000) or 
(monthname(birth_date) not in ('July','August','September'))

select *
from orders
where shipped_date is not null

-- Text Operator 
-- like operator 
-- 'be%'
-- '%ey'
-- 'bo%ey'
-- 'm__c%' 
-- '%m__c' 
 


select *
from customers
where last_name like '%e_y'
  
 
-- regexp
-- '^a'
-- 'y$' 
-- 'a[g-n][q,r,s]$'
-- 'ey'  
-- '^ac|ey$' 


select *
from customers
where last_name regexp 'ey' 

-- Relational Operator(<,>,=,!=,>=,<=) 
-- Logical Operator(and,or, not)
-- betweeen 
-- membership operator (in , not in)
-- identity operator( is , is not)
-- like
-- regexp 

select * 
from clients
where client_id in(
					select client_id
					from payments
					where payment_method = (
											select payment_method_id
											from payment_methods 
											where name = "credit card"))

```
