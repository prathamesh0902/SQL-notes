## Implementing Linear Regression

```python
import numpy as np
```

```python
n1 = []
n2 = []

for i in range(1,100):
    n1.append(i*np.random.randint(1, 10))
    n2.append(i+np.random.randint(1, 10)+i**2)

n1 = np.array(n1)
n2 = np.array(n2)
# df = spark.createDataFrame(zip(n1, n2), ['a', 'b'])
```

```python
w0 = round(np.random.random(),2)
w1 = round(np.random.random(),2)
i = 1
alpha = 0.001

while i <= 50:
    yp = w0 + w1*n1
    j = sum((yp-n2)**2)/len(n1)
    print(j)
    if j > 0 and j < 1:
        break
    dw0 = sum(yp-n2)/len(n1)
    dw1 = sum((yp-n2)*n1)/len(n1)
    w0 = w0 - alpha * dw0
    w1 = w1 - alpha * dw1
    i += 1
```
