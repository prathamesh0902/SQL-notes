## DCL (Data Control Language)
`grant` `revoke`

```sql
create user 'Adam' identified by '123';

grant select, insert, update, delete on database default to user 'Adam';

grant select, insert, update, delete on sql_store.* to user 'Adam';
```

```sql
revoke update, insert on sql_store.customers from user 'Adam';

drop user 'Adam';
```

## TCL (Transaction Control Language)
`rollback` `commit` `savepoint`

```sql
-- add content from class

start transaction; -- to enable transactions

use sql_store
insert into products
values(default, "Moov", 2.1, 10),
(default, "Monocef", 3.1, 12),
(default, "PCM",4.1, 15)

commit; -- to commit the changes
```

```sql
start transaction;

update products
set unit_price = unit_price * 1.05
where product_id= 5;

delete from products
where product_id = 11;

rollback; -- to the last commit point
```

```sql
start transaction;

savepoint a;

use sql_store
insert into products
values(default, "Moov", 2.1, 10),
(default, "Monocef", 3.1, 12),
(default, "PCM",4.1, 15)

savepoint b;
update products
set unit_price = unit_price * 1.05
where product_id= 5;

savepoint c;
delete from products
where product_id = 11;

rollback to b; -- to the savepoint
```
