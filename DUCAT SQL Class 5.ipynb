## Data Definition Language (DDL)

`create` `alter` `drop` `truncate`

```sql
alter table customers
add column Emergency_number varchar(14) not null;

alter table customers
drop column Emergency_number;

alter table customers
add column Emergency_number varchar(14) not null after phone;

alter table customers
rename column birth_date to dob

alter table customers
modify column state varchar(20) not null;



-- employees child to offices;
drop table employees;

drop table offices;

drop database sql_hr;


-- truncate erases all the data within
truncate table products
```

`key` concept

- set of attributes that help us uniquely identify the data
- force identity of data to ensure integrity
- to establish and identify relation between tables

1. Super key - uniqely identifiable, every key in the table
2. Candidate key - minimal subset of super key
3. Primary key - candidate key with no duplicates, no nulls
4. Foreign key
5. Composite and Compound key


Creating Database
- Make the ER diagram

```sql
create database hospital_db;
```

```sql
use hospital_db;

create table department (
    dept_id int not null,
    dept_name varchar(20) not null,
    established date not null,
    loc varchar(20) default 'NA',
    PRIMARY KEY(dept_id) -- primary key(c1, c2)
);
```

```sql
create table employees(
  emp_id int not null auto_increment primary key,
  emp_name varchar(50) not null,
  designation varchar(50) not null,
  dept_id int not null,
  phone varchar(14) not null,
  foreign key(dept_id) references department(dept_id)

)
```
